## Exemplo Gr√°fico de Linha

```python
import matplotlib.pyplot as plt

# Dias da semana e temperaturas correspondentes
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "S√°b", "Dom"]
temperaturas = [28, 30, 27, 25, 26, 29, 31]

# Cria√ß√£o do gr√°fico de linha
plt.plot(dias, temperaturas, marker='o', linestyle='-', color='blue')  # Linha azul com marcadores
plt.title("Varia√ß√£o da Temperatura Durante a Semana")  # T√≠tulo do gr√°fico
plt.xlabel("Dia da Semana")  # R√≥tulo eixo X
plt.ylabel("Temperatura (¬∞C)")  # R√≥tulo eixo Y
plt.grid(True)  # Grade no gr√°fico
plt.tight_layout()  # Ajusta layout
plt.show()  # Exibe o gr√°fico
```

```python
import matplotlib.pyplot as plt

# Meses e faturamento em R$
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"]
faturamento = [12000, 13500, 12800, 14500, 15000, 16000]

# Gr√°fico de linha para mostrar evolu√ß√£o
plt.plot(meses, faturamento, marker='s', linestyle='--', color='green')  # Linha verde tracejada com quadrados
plt.title("Faturamento Mensal da Empresa")  # T√≠tulo
plt.xlabel("M√™s")  # Eixo X
plt.ylabel("Faturamento (R$)")  # Eixo Y
plt.grid(True, linestyle=':', alpha=0.6)  # Grade personalizada
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Provas e respectivas notas
provas = ["Prova 1", "Prova 2", "Prova 3", "Prova 4", "Prova 5"]
notas = [60, 75, 70, 85, 90]

# Cria√ß√£o do gr√°fico de linha
plt.plot(provas, notas, marker='^', linestyle='-', color='purple')  # Tri√¢ngulo como marcador
plt.title("Desempenho do Estudante ao Longo das Avalia√ß√µes")  # T√≠tulo
plt.xlabel("Avalia√ß√£o")  # Eixo X
plt.ylabel("Nota")  # Eixo Y
plt.ylim(0, 100)  # Define limite do eixo Y
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista com os nomes das avalia√ß√µes aplicadas
provas = ["Prova 1", "Prova 2", "Prova 3", "Prova 4", "Prova 5"]

# Notas dos alunos ao longo das 5 avalia√ß√µes
aluno_A = [70, 75, 80, 85, 90]  # Aluno A teve desempenho crescente
aluno_B = [60, 65, 68, 70, 72]  # Aluno B com crescimento mais lento
aluno_C = [50, 60, 55, 65, 80]  # Aluno C com desempenho irregular

# Cria√ß√£o da figura e defini√ß√£o do tamanho
plt.figure(figsize=(8, 5))  # Tamanho do gr√°fico (largura x altura)

# Linha representando o Aluno A
plt.plot(provas, aluno_A, 
         marker='o',          # Marcador circular
         linestyle='-',      # Linha cont√≠nua
         color='blue',       # Cor azul
         label='Aluno A')    # R√≥tulo para legenda

# Linha representando o Aluno B
plt.plot(provas, aluno_B, 
         marker='s',         # Marcador quadrado
         linestyle='--',     # Linha tracejada
         color='green',      # Cor verde
         label='Aluno B')    # R√≥tulo para legenda

# Linha representando o Aluno C
plt.plot(provas, aluno_C, 
         marker='^',         # Marcador triangular
         linestyle='-.',     # Linha tra√ßo-ponto
         color='red',        # Cor vermelha
         label='Aluno C')    # R√≥tulo para legenda

# T√≠tulo do gr√°fico
plt.title("Compara√ß√£o de Desempenho de Tr√™s Alunos")

# R√≥tulos dos eixos
plt.xlabel("Avalia√ß√µes")     # Nome do eixo X
plt.ylabel("Nota")           # Nome do eixo Y

# Define o limite do eixo Y entre 40 e 100
plt.ylim(40, 100)

# Adiciona linhas de grade ao fundo do gr√°fico
plt.grid(True, linestyle=':', alpha=0.7)

# Exibe a legenda no canto inferior direito
plt.legend(loc='lower right')

# Ajusta automaticamente os elementos do layout para n√£o sobrepor
plt.tight_layout()

# Exibe o gr√°fico
plt.show()
```
