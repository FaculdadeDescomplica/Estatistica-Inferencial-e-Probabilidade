## Exemplo Gráfico de Linha

```python
import matplotlib.pyplot as plt

# Dias da semana e temperaturas correspondentes
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"]
temperaturas = [28, 30, 27, 25, 26, 29, 31]

# Criação do gráfico de linha
plt.plot(dias, temperaturas, marker='o', linestyle='-', color='blue')  # Linha azul com marcadores
plt.title("Variação da Temperatura Durante a Semana")  # Título do gráfico
plt.xlabel("Dia da Semana")  # Rótulo eixo X
plt.ylabel("Temperatura (°C)")  # Rótulo eixo Y
plt.grid(True)  # Grade no gráfico
plt.tight_layout()  # Ajusta layout
plt.show()  # Exibe o gráfico
```

```python
import matplotlib.pyplot as plt

# Meses e faturamento em R$
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"]
faturamento = [12000, 13500, 12800, 14500, 15000, 16000]

# Gráfico de linha para mostrar evolução
plt.plot(meses, faturamento, marker='s', linestyle='--', color='green')  # Linha verde tracejada com quadrados
plt.title("Faturamento Mensal da Empresa")  # Título
plt.xlabel("Mês")  # Eixo X
plt.ylabel("Faturamento (R$)")  # Eixo Y
plt.grid(True, linestyle=':', alpha=0.6)  # Grade personalizada
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Provas e respectivas notas
provas = ["Prova 1", "Prova 2", "Prova 3", "Prova 4", "Prova 5"]
notas = [60, 75, 70, 85, 90]

# Criação do gráfico de linha
plt.plot(provas, notas, marker='^', linestyle='-', color='purple')  # Triângulo como marcador
plt.title("Desempenho do Estudante ao Longo das Avaliações")  # Título
plt.xlabel("Avaliação")  # Eixo X
plt.ylabel("Nota")  # Eixo Y
plt.ylim(0, 100)  # Define limite do eixo Y
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista com os nomes das avaliações aplicadas
provas = ["Prova 1", "Prova 2", "Prova 3", "Prova 4", "Prova 5"]

# Notas dos alunos ao longo das 5 avaliações
aluno_A = [70, 75, 80, 85, 90]  # Aluno A teve desempenho crescente
aluno_B = [60, 65, 68, 70, 72]  # Aluno B com crescimento mais lento
aluno_C = [50, 60, 55, 65, 80]  # Aluno C com desempenho irregular

# Criação da figura e definição do tamanho
plt.figure(figsize=(8, 5))  # Tamanho do gráfico (largura x altura)

# Linha representando o Aluno A
plt.plot(provas, aluno_A, 
         marker='o',          # Marcador circular
         linestyle='-',      # Linha contínua
         color='blue',       # Cor azul
         label='Aluno A')    # Rótulo para legenda

# Linha representando o Aluno B
plt.plot(provas, aluno_B, 
         marker='s',         # Marcador quadrado
         linestyle='--',     # Linha tracejada
         color='green',      # Cor verde
         label='Aluno B')    # Rótulo para legenda

# Linha representando o Aluno C
plt.plot(provas, aluno_C, 
         marker='^',         # Marcador triangular
         linestyle='-.',     # Linha traço-ponto
         color='red',        # Cor vermelha
         label='Aluno C')    # Rótulo para legenda

# Título do gráfico
plt.title("Comparação de Desempenho de Três Alunos")

# Rótulos dos eixos
plt.xlabel("Avaliações")     # Nome do eixo X
plt.ylabel("Nota")           # Nome do eixo Y

# Define o limite do eixo Y entre 40 e 100
plt.ylim(40, 100)

# Adiciona linhas de grade ao fundo do gráfico
plt.grid(True, linestyle=':', alpha=0.7)

# Exibe a legenda no canto inferior direito
plt.legend(loc='lower right')

# Ajusta automaticamente os elementos do layout para não sobrepor
plt.tight_layout()

# Exibe o gráfico
plt.show()
```
