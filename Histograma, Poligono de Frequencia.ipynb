## Exemplo 1 - Amostra com 30 numeros

```python
# Importa a biblioteca matplotlib.pyplot e atribui o apelido 'plt'
# Essa biblioteca é usada para criar gráficos e visualizações em Python
import matplotlib.pyplot as plt

# Lista com as notas dos alunos
notas = [45, 56, 78, 90, 65, 70, 85, 60, 55, 72, 68, 80, 74, 88, 92,
         67, 64, 59, 50, 73, 77, 69, 61, 66, 54, 75, 62, 58, 83, 79]

# Cria o histograma com os dados
plt.hist(notas, bins=6, edgecolor='black', alpha=0.7)
# 'bins=6' divide os dados em 6 intervalos (faixas de valores)
# 'edgecolor="black"' desenha a borda preta em cada barra
# 'alpha=0.7' define a transparência das barras (0 = totalmente transparente, 1 = opaco)

plt.title('Histograma das Notas') # Define o título do gráfico
plt.xlabel('Notas') # Define o rótulo do eixo X
plt.ylabel('Frequência') # Define o rótulo do eixo Y

# Exibe o gráfico na tela
plt.show()
```

```python
# Importa a biblioteca NumPy, usada para cálculos numéricos
import numpy as np

# Lista com as notas dos alunos
notas = [45, 56, 78, 90, 65,
         70, 85, 60, 55, 72, 68, 80, 74, 88, 92,         
         67, 64, 59, 50, 73, 77, 69, 61, 66, 54, 75, 62, 58, 83, 79]   

# Gera os dados do histograma, mas sem plotar:
hist, bins = np.histogram(notas, bins=6)  # bins=6 define 6 intervalos de classe
# 'hist' contém as frequências de cada intervalo (bin)
# 'bins' contém os limites dos intervalos (bordas dos bins)

# Calcula os pontos médios de cada intervalo
mid_bins = 0.5 * (bins[:-1] + bins[1:]) ## Isso é necessário para plotar o gráfico de linha (polígono)

# Cria o polígono de frequências:
plt.plot(mid_bins, hist, marker='o', linestyle='-', color='b', alpha=0.7)
# - 'mid_bins' é o eixo x (centros dos intervalos)
# - 'hist' é o eixo y (frequências absolutas)
# - 'marker="o"' coloca marcadores nos pontos
# - 'linestyle="-"' conecta os pontos com linhas
# - 'color="b"' define a cor azul
# - 'alpha=0.7' define transparência

plt.title('Polígono de Frequências das Notas') # Define o título do gráfico

# Define os rótulos dos eixos
plt.xlabel('Notas') 
plt.ylabel('Frequência')

# Exibe o gráfico
plt.show()
```
