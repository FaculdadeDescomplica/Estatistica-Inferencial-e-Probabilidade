```python
# Importa bibliotecas necessárias
import numpy as np          # Para cálculos e geração de números aleatórios
import matplotlib.pyplot as plt  # Para criação de gráficos
```

```python
# --- 1. Definição do parâmetro da distribuição ---
media = 20          # Tempo médio entre atendimentos em minutos
lambda_ = 1 / media  # Taxa λ da distribuição exponencial
```

```python
# --- 2. Cálculo de probabilidade ---
tempo = 30  # Queremos saber a probabilidade de atendimento durar mais de 30 minutos
# Fórmula da exponencial: P(X > x) = e^(-λx)
prob_mais_30 = np.exp(-lambda_ * tempo)
print(f"Probabilidade de atendimento > 30 min: {prob_mais_30:.4f}")
```

```python
# --- 3. Simulação de dados ---
# Geramos 1000 tempos de atendimento aleatórios seguindo a distribuição exponencial
tempos_simulados = np.random.exponential(scale=media, size=1000)
```

```python
# --- 4. Criação do histograma ---
# Plota a distribuição dos tempos simulados
plt.hist(
    tempos_simulados,   # dados simulados
    bins=30,           # número de "caixas" do histograma
    color='skyblue',   # cor das barras
    edgecolor='black', # cor da borda das barras
    density=True       # normaliza para densidade de probabilidade
)
plt.title("Distribuição Exponencial - Tempo de Atendimento")  # título do gráfico
plt.xlabel("Tempo de Atendimento (minutos)")  # legenda eixo x
plt.ylabel("Densidade")                      # legenda eixo y
plt.show()                                   # exibe o gráfico
```
