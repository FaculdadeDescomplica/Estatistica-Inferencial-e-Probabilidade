Em uma col√¥nia de 15 bact√©rias, a chance de uma bact√©ria ser resistente a antibi√≥tico √©  ùëù=0,2. 
Qual a probabilidade de exatamente 3 bact√©rias serem resistentes?

```python
# Importando bibliotecas necess√°rias
import matplotlib.pyplot as plt  # Para criar gr√°ficos
from scipy.stats import binom    # Para usar a distribui√ß√£o binomial
import numpy as np               # Para criar arrays de valores de k
```

```python
# --- Par√¢metros do problema ---
n = 15        # N√∫mero total de bact√©rias (n√∫mero de tentativas)
p = 0.2       # Probabilidade de uma bact√©ria ser resistente (sucesso)
k = 3         # N√∫mero de bact√©rias resistentes que queremos calcular a probabilidade
```

```python
# --- C√°lculo da probabilidade exata de k sucessos ---

# A fun√ß√£o binom.pmf(k, n, p) calcula P(X = k) para a distribui√ß√£o binomial
probabilidade_exata = binom.pmf(k, n, p)

print(f"A probabilidade de exatamente {k} bact√©rias serem resistentes √© {probabilidade_exata:.3f}")
```

```python
# --- Valores poss√≠veis de k para a distribui√ß√£o ---
# Criamos um array de 0 at√© n, representando todas as quantidades poss√≠veis de bact√©rias resistentes
k_values = np.arange(0, n)

# --- Calculando a probabilidade para cada valor de k ---
probabilidades = binom.pmf(k_values, n, p)

# --- Criando o gr√°fico da distribui√ß√£o binomial ---
plt.figure(figsize=(10,6))  # Define o tamanho do gr√°fico
plt.bar(k_values, probabilidades, color='skyblue', edgecolor='black')  # Barras azuis para todas as probabilidades
plt.xlabel('N√∫mero de bact√©rias resistentes (k)')  # Label eixo X
plt.ylabel('Probabilidade P(X=k)')                 # Label eixo Y
plt.title('Distribui√ß√£o Binomial - Bact√©rias resistentes')  # T√≠tulo do gr√°fico
plt.xticks(k_values)  # Mostra todos os valores de k no eixo X

# --- Destacar o valor k=3 ---
# Pintamos a barra de k=3 em laranja para enfatizar a probabilidade calculada
plt.bar(k, binom.pmf(k, n, p), color='orange', edgecolor='black', label=f'k={k}')

plt.legend()  # Adiciona legenda
plt.show()    # Exibe o gr√°fico
```
