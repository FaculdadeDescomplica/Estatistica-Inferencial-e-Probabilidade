## Exemplo de BOX-PLOT

```python
import matplotlib.pyplot as plt

# Lista com as notas dos alunos
notas = [45, 50, 52, 53, 55, 60, 65, 68, 70, 72, 75, 80, 85, 90, 95]

# Criação do gráfico Box Plot
plt.boxplot(notas, patch_artist=True, boxprops=dict(facecolor='lightgreen'))  # Cor personalizada
plt.title("Box Plot - Notas dos Alunos")  # Título do gráfico
plt.ylabel("Notas")  # Rótulo do eixo Y
plt.grid(True)  # Adiciona grade ao gráfico
plt.show()  # Exibe o gráfico

```

```python
import matplotlib.pyplot as plt

# Lista de salários mensais (um valor atípico: 10000)
salarios = [2000, 2200, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 10000]

# Criação do Box Plot com visual personalizado
plt.boxplot(salarios, patch_artist=True, boxprops=dict(facecolor='salmon'))  # Cor salmão
plt.title("Box Plot - Salários")  # Título do gráfico
plt.ylabel("Salário (R$)")  # Rótulo do eixo Y
plt.grid(True)  # Adiciona linhas de grade
plt.show()  # Mostra o gráfico na tela
```

```python
import matplotlib.pyplot as plt

# Notas de duas turmas distintas
turma_A = [55, 60, 65, 70, 72, 75, 78, 80, 83, 85]
turma_B = [45, 50, 55, 58, 60, 62, 64, 66, 68, 70]

# Criação de um gráfico comparando as duas distribuições
plt.boxplot([turma_A, turma_B],  # Lista de listas
            patch_artist=True,  # Preenche a caixa com cor
            boxprops=dict(facecolor='lightblue'),  # Define cor das caixas
            labels=["Turma A", "Turma B"])  # Rótulos de cada boxplot

plt.title("Box Plot - Comparação de Notas entre Turmas")  # Título do gráfico
plt.ylabel("Notas")  # Eixo Y
plt.grid(True)  # Adiciona grade
plt.show()  # Exibe o gráfico
```
