## 3º Um departamento realiza auditorias em documentos confidenciais. Em cada auditoria, são selecionados aleatoriamente n documentos de um total N, dos quais K são confidenciais.
## Calcular e visualizar a probabilidade de diferentes números de documentos confidenciais serem escolhidos na amostra, considerando três cenários diferentes: 

# Cenário 1: N=80, K=12, n=8
# Cenário 2: N=100, K=20,n=10
# Cenário 3: N=60, K=15, n=5


```python
from math import comb        # Importa a função 'comb' para calcular coeficiente binomial
import matplotlib.pyplot as plt  # Importa a biblioteca para criação de gráficos
```

```python
def hipergeometrica_visual(N, K, n):
    """
    Calcula e plota a distribuição hipergeométrica para qualquer combinação de parâmetros.
    
    Parâmetros:
    N : int - tamanho da população total
    K : int - número de sucessos na população
    n : int - tamanho da amostra retirada
    """
    
    # Calcula a probabilidade para cada número de sucessos na amostra (de 0 a n)
    # Fórmula: P(X=k) = (C(K, k) * C(N-K, n-k)) / C(N, n)
    probabilidades = [ (comb(K, k) * comb(N-K, n-k)) / comb(N, n) for k in range(n+1) ]
    
    # Cria o gráfico de barras para visualizar a distribuição
    plt.figure(figsize=(8,5))  # Define o tamanho da figura
    plt.bar(range(n+1), probabilidades, color='skyblue')  # Cria barras para cada k
    plt.xticks(range(n+1))      # Define os rótulos do eixo x de 0 a n
    plt.xlabel("Número de sucessos na amostra")  # Label do eixo x
    plt.ylabel("Probabilidade")                # Label do eixo y
    plt.title(f"Distribuição Hipergeométrica (N={N}, K={K}, n={n})")  # Título do gráfico
    plt.show()  # Exibe o gráfico
    
    # Retorna a lista de probabilidades calculadas
    return probabilidades
```

```python
# Definindo três cenários diferentes para análise
cenarios = [
    {"N": 80, "K": 12, "n": 8},   # Cenário 1
    {"N": 100, "K": 20, "n": 10}, # Cenário 2
    {"N": 60, "K": 15, "n": 5}    # Cenário 3
]
```

```python
# Laço para calcular e exibir resultados de cada cenário
for i, c in enumerate(cenarios, start=1):
    print(f"\nCenário {i}: N={c['N']}, K={c['K']}, n={c['n']}")
    
    # Chama a função para calcular e plotar as probabilidades
    probs = hipergeometrica_visual(c['N'], c['K'], c['n'])
    
    # Exibe no console a probabilidade de cada número de sucessos na amostra
    for k, p in enumerate(probs):
        print(f"Probabilidade de {k} sucessos: {p:.4f}")
```
