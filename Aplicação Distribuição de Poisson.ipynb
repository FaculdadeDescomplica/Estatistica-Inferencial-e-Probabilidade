## Questão — Controle de Qualidade em Produção ##

Em uma fábrica, a média de defeitos por lote é 2 defeitos por lote. Suponha que o número de defeitos siga uma Distribuição de Poisson com λ = 2 .

a) Calcule a probabilidade de nenhum defeito em um lote.

b) Calcule a probabilidade de exatamente 1 defeito.

c) Calcule a probabilidade de 3 ou mais defeitos.

d) Gere um gráfico da Distribuição de Poisson para valores de 0 a 6 defeitos, destacando visualmente os resultados dos itens 1 a 3.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# --- Parâmetro da distribuição ---
lambda_defects = 2  # média de defeitos por lote (λ)

# --- Valores possíveis de defeitos ---
k_values = np.arange(0, 7)  # consideramos 0 a 6 defeitos

# --- Probabilidades de cada número de defeitos ---
probs = poisson.pmf(k_values, lambda_defects)  # função de massa de probabilidade Poisson
```

```python
# --- Cálculos solicitados ---
P_0 = poisson.pmf(0, lambda_defects)          # probabilidade de 0 defeitos
P_1 = poisson.pmf(1, lambda_defects)          # probabilidade de 1 defeito
P_3_or_more = 1 - poisson.cdf(2, lambda_defects)  # probabilidade de 3 ou mais defeitos

# --- Mostrar os resultados no console ---
print(f"Probabilidade de 0 defeitos: {P_0:.4f} ({P_0*100:.2f}%)")
print(f"Probabilidade de 1 defeito: {P_1:.4f} ({P_1*100:.2f}%)")
print(f"Probabilidade de 3 ou mais defeitos: {P_3_or_more:.4f} ({P_3_or_more*100:.2f}%)")
```

```python
# --- Gráfico ---
plt.figure(figsize=(10,6))  # define o tamanho da figura
plt.bar(k_values, probs, color='skyblue', edgecolor='black', label='Probabilidade Poisson')  # histograma principal

# Destacar os itens solicitados em cores diferentes
plt.bar(0, P_0, color='orange', label='0 defeitos')   # destaca 0 defeitos
plt.bar(1, P_1, color='green', label='1 defeito')    # destaca 1 defeito
plt.bar(3, probs[3], color='red', label='≥3 defeitos (primeiro valor)')  # destaca ≥3 defeitos (o primeiro valor da área)

# --- Configurações do gráfico ---
plt.title("Distribuição de Poisson - Defeitos por Lote")  # título
plt.xlabel("Número de defeitos")  # eixo X
plt.ylabel("Probabilidade")       # eixo Y
plt.xticks(k_values)              # marcar cada valor de 0 a 6
plt.legend()                      # mostrar legenda
plt.grid(True, linestyle='--', alpha=0.5)  # grade tracejada para facilitar leitura
plt.show()                        # exibir gráfico
```
