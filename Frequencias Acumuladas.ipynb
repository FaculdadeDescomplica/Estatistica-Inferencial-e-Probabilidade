## Exemplo 1  - Amostra com 100 elementos


```python
import pandas as pd                    # Importa o pandas para manipulação de dados em tabelas
import matplotlib.pyplot as plt        # Importa matplotlib para plotar gráficos
```

```python
# Exemplo de dados
dados = [3, 10, 2, 9, 8, 4, 6, 8, 5, 10, 3, 2, 4, 10, 3, 
         4, 8, 2, 2, 1, 2, 5, 8, 8, 2, 4, 9, 10, 5, 5, 8, 
         3, 1, 9, 8, 1, 3, 6, 5, 9, 2, 2, 2, 9, 5, 9, 6, 5,
         8, 3, 3, 5, 10, 2, 4, 9, 9, 8, 4, 3, 3, 6, 10, 3, 
         4, 1, 4, 3, 3, 10, 5, 10, 6, 5, 9, 2, 5, 4, 2, 8, 
         10, 1, 4, 2, 10, 5, 4, 5, 6, 3, 6, 1, 9, 8, 10, 10,
         10, 2, 2, 1, 3, 9, 5, 4, 1, 5, 9, 4, 1, 10, 4, 3,
         3, 9, 4, 9, 4, 2, 2, 7, 7, 6, 8, 2, 3, 4, 5, 10,
         2, 6, 3, 4, 5, 8]
```

```python
# Converte a lista em um DataFrame do pandas, nomeando a coluna como 'Valor'
df = pd.DataFrame(dados, columns=['Valor'])
```

```python
# Calcula a frequência absoluta (fi) - quantas vezes cada valor aparece
freq_abs = df['Valor'].value_counts().sort_index()  
# .value_counts() conta os valores
# .sort_index() ordena pelo valor dos dados (ex: 1,2,3,...)

# Calcula a frequência absoluta acumulada (Fi acumulada)
freq_abs_acum = freq_abs.cumsum()  
# .cumsum() soma cumulativa, acumulando as frequências até cada valor

# Calcula a frequência relativa (fr) - proporção do total de cada valor
freq_rel = freq_abs / freq_abs.sum()  
# divide cada frequência absoluta pela soma total das frequências

# Calcula a frequência relativa acumulada (Fr acumulada)
freq_rel_acum = freq_rel.cumsum()  
# soma cumulativa das frequências relativas

# Calcula a frequência percentual acumulada (Fp acumulada)
freq_perc_acum = freq_rel_acum * 100  
# converte a frequência relativa acumulada para porcentagem
```

```python
# Junta todos os resultados em um único DataFrame para visualização organizada
tabela_freq = pd.DataFrame({
    'Frequência Absoluta': freq_abs,
    'Frequência Absoluta Acumulada': freq_abs_acum,
    'Frequência Relativa': freq_rel,
    'Frequência Relativa Acumulada': freq_rel_acum,
    'Frequência Percentual Acumulada (%)': freq_perc_acum
})

# Exibe a tabela com todas as frequências calculadas
tabela_freq

```

```python
# --- Visualização gráfica da frequência absoluta acumulada ---

plt.figure(figsize=(8,5))  # Define o tamanho da figura do gráfico
plt.step(tabela_freq.index, tabela_freq['Frequência Absoluta Acumulada'], 
         where='mid', label='Fi Acumulada')  # Gráfico em degraus (step plot) da Fi acumulada
plt.plot(tabela_freq.index, tabela_freq['Frequência Absoluta Acumulada'], 'o')  # Pontos nos valores exatos
plt.title('Frequência Absoluta Acumulada')  # Título do gráfico
plt.xlabel('Valores')  # Legenda do eixo x
plt.ylabel('Frequência Absoluta Acumulada')  # Legenda do eixo y
plt.grid(True)  # Adiciona grid para facilitar visualização
plt.legend()  # Mostra a legenda
plt.show()  # Exibe o gráfico
```

```python
# --- Visualização gráfica da frequência percentual acumulada ---
plt.figure(figsize=(8,5))  # Define o tamanho da figura do gráfico
plt.step(tabela_freq.index, tabela_freq['Frequência Percentual Acumulada (%)'], 
         where='mid', label='Fp Acumulada (%)', color='purple')  # Gráfico step da frequência percentual acumulada
plt.plot(tabela_freq.index, tabela_freq['Frequência Percentual Acumulada (%)'], 'o', color='purple')  # Pontos
plt.title('Frequência Percentual Acumulada (%)')  # Título do gráfico
plt.xlabel('Valores')  # Legenda do eixo x
plt.ylabel('Frequência Percentual Acumulada (%)')  # Legenda do eixo y
plt.grid(True)  # Adiciona grid
plt.legend()  # Exibe legenda
plt.show()  # Exibe gráfico
```

## Exemplo 2  - Amostra com 1000 elementos


```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
import numpy as np

# Gera 1000 números seguindo distribuição normal
dados_float = np.random.normal(loc=5.5, scale=3, size=1000)

# Arredonda para inteiro e limita entre 1 e 10
dados = np.clip(np.round(dados_float), 1, 10).astype(int)

print(dados)
```

```python
# Criar tabela de frequências
df = pd.DataFrame(dados, columns=['Valor'])
```

```python
# Frequência absoluta (fi)
freq_abs = df['Valor'].value_counts().sort_index()

# Frequência absoluta acumulada (Fi acumulada)
freq_abs_acum = freq_abs.cumsum()

# Frequência relativa (fr)
freq_rel = freq_abs / freq_abs.sum()

# Frequência relativa acumulada (Fr acumulada)
freq_rel_acum = freq_rel.cumsum()

# Frequência percentual acumulada (Fp acumulada)
freq_perc_acum = freq_rel_acum * 100
```

```python
# Montar DataFrame com todas as frequências
tabela_freq = pd.DataFrame({
    'Frequência Absoluta': freq_abs,
    'Frequência Absoluta Acumulada': freq_abs_acum,
    'Frequência Relativa': freq_rel,
    'Frequência Relativa Acumulada': freq_rel_acum,
    'Frequência Percentual Acumulada (%)': freq_perc_acum
})

tabela_freq
```

```python
# Visualização gráfica da frequência absoluta acumulada
plt.figure(figsize=(8,5))
plt.step(tabela_freq.index, tabela_freq['Frequência Absoluta Acumulada'], where='mid', label='Fi Acumulada')
plt.plot(tabela_freq.index, tabela_freq['Frequência Absoluta Acumulada'], 'o')
plt.title('Frequência Absoluta Acumulada')
plt.xlabel('Valores')
plt.ylabel('Frequência Absoluta Acumulada')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Gráfico da frequência percentual acumulada
plt.figure(figsize=(8,5))
plt.step(tabela_freq.index, tabela_freq['Frequência Percentual Acumulada (%)'], where='mid', label='Fp Acumulada (%)', color='purple')
plt.plot(tabela_freq.index, tabela_freq['Frequência Percentual Acumulada (%)'], 'o', color='purple')
plt.title('Frequência Percentual Acumulada (%)')
plt.xlabel('Valores')
plt.ylabel('Frequência Percentual Acumulada (%)')
plt.grid(True)
plt.legend()
plt.show()
```
