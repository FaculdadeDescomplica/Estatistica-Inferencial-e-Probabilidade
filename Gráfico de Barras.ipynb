# Exemplo Gr√°fico de Barras

```python
import matplotlib.pyplot as plt  # Importa a biblioteca para cria√ß√£o de gr√°ficos

# Dados: categorias e valores
categorias = ['A', 'B', 'C', 'D']
valores = [10, 25, 15, 30]

plt.bar(categorias, valores, color='orange')  # Cria barras verticais com cor laranja
plt.title('Gr√°fico de Barras Simples')        # Adiciona t√≠tulo ao gr√°fico
plt.xlabel('Categorias')                      # Nomeia o eixo x
plt.ylabel('Valores')                         # Nomeia o eixo y
plt.show()                                   # Exibe o gr√°fico

```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
frutas = ['Ma√ß√£', 'Banana', 'Laranja', 'Uva']
quantidades = [40, 55, 30, 25]

plt.barh(frutas, quantidades, color=['red', 'yellow', 'orange', 'purple'])  # Barras horizontais coloridas
plt.title('Quantidade de Frutas Vendidas')                                # T√≠tulo do gr√°fico
plt.xlabel('Quantidade')                                                   # Nome do eixo x
plt.ylabel('Frutas')                                                      # Nome do eixo y
plt.show()                                                                # Exibe o gr√°fico
```

```python
import matplotlib.pyplot as plt

# Dados: g√™neros liter√°rios e vendas
generos = ['Romance', 'Fantasia', 'Fic√ß√£o Cient√≠fica', 'Autoajuda']
vendas = [120, 95, 80, 70]

plt.bar(generos, vendas, color='skyblue')  # Gr√°fico de barras vertical com cor azul clara
plt.title('Vendas por G√™nero Liter√°rio')   # T√≠tulo do gr√°fico
plt.xlabel('G√™nero Liter√°rio')             # R√≥tulo do eixo x
plt.ylabel('Quantidade Vendida')           # R√≥tulo do eixo y

# Loop para mostrar os valores no topo das barras
for i, venda in enumerate(vendas):
    plt.text(i, venda + 2, str(venda), ha='center')

plt.show()  # Mostra o gr√°fico
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Categorias no eixo X
categorias = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril']

# Dados para dois grupos (ex: vendas em duas lojas diferentes)
vendas_loja1 = [150, 200, 180, 220]
vendas_loja2 = [170, 190, 160, 210]

# Define a posi√ß√£o das barras no eixo X
x = np.arange(len(categorias))  # array: [0,1,2,3]
largura = 0.35  # largura das barras

plt.figure(figsize=(10,6))  # tamanho da figura

# Barras da Loja 1 (posi√ß√£o x)
barra1 = plt.bar(x - largura/2, vendas_loja1, largura, label='Loja 1', color='cornflowerblue')

# Barras da Loja 2 (posi√ß√£o x deslocada para direita)
barra2 = plt.bar(x + largura/2, vendas_loja2, largura, label='Loja 2', color='salmon')

# Adiciona t√≠tulo e r√≥tulos
plt.title('Vendas Mensais por Loja', fontsize=16)
plt.xlabel('M√™s', fontsize=14)
plt.ylabel('Quantidade de Vendas', fontsize=14)
plt.xticks(x, categorias)  # substitui valores do eixo x por nomes dos meses
plt.legend()  # mostra legenda para identificar as lojas

# Fun√ß√£o para adicionar r√≥tulo de valores no topo das barras
def adicionar_valores(barras):
    for barra in barras:
        altura = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2, altura + 5, 
                 str(altura), ha='center', fontsize=10)

adicionar_valores(barra1)
adicionar_valores(barra2)

plt.tight_layout()  # ajusta o layout para n√£o cortar elementos
plt.show()
```
